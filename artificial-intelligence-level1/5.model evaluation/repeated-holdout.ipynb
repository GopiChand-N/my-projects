```python
import pandas as pd
import os
from sklearn import tree 
from sklearn import preprocessing
from sklearn import model_selection
import numpy as np
```

```python
dir = "C:/Users/pc/Downloads/ai-level1/titanic"
titanic_train = pd.read_csv(os.path.join(dir, "train.csv"))
titanic_train.shape
titanic_train.info()
```

```python
le_sex = preprocessing.LabelEncoder()
le_sex.fit(titanic_train['Sex'])
print(le_sex.classes_)
titanic_train['Sex_encoded'] = le_sex.transform(titanic_train['Sex'])
titanic_train.head(20)
```

```python
le_emb = preprocessing.LabelEncoder()
le_emb.fit(titanic_train['Embarked'])
print(le_emb.classes_)
titanic_train['Embarked_encoded'] = le_emb.transform(titanic_train['Embarked'])
titanic_train.head(20)
```

```python
dt = tree.DecisionTreeClassifier()
features = ['Parch', 'SibSp', 'Sex_encoded', 'Embarked_encoded']
X_train = titanic_train[features]
y_train = titanic_train['Survived']
dt.fit(X_train, y_train)
```

```python
%matplotlib qt
tree.plot_tree(dt, feature_names=features, filled=True, fontsize=7)
```

```python
ss = model_selection.ShuffleSplit(10, train_size=0.7)
splits = ss.split(X_train)
print(splits)
for t, v in splits:
    print(t, v)
    #print(X_train[t], y_train[t])
    #print(X_train[v], y_train[v])
```

```python
scores = model_selection.cross_val_score(dt, X_train, y_train, scoring='accuracy', cv=ss)
print(scores, scores.mean())
```

```python
titanic_test = pd.read_csv(os.path.join(dir, "test.csv"))
print(titanic_test.shape)
titanic_test.head()
```

```python
titanic_test['Sex_encoded'] = le_sex.transform(titanic_test['Sex'])
titanic_test['Embarked_encoded'] = le_emb.transform(titanic_test['Embarked'])
X_test = titanic_test[features]
titanic_test['Survived'] = dt.predict(X_test)
titanic_test.head()
```

```python
titanic_test.to_csv(os.path.join(dir, "submission1.csv"), columns=["PassengerId", "Survived"], index=False)
```
